**5. Show the plot of the listings on a map**

```python
affordable_df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# menampilkan lokasi berdasarkan neighbourhood group
fig = plt.subplots(figsize=(10, 10))
sns.scatterplot(data=affordable_df, x='longitude',y='latitude', hue='neighbourhood_group', palette='Set2').set_title('Lokasi ', size=20)
plt.show()
```

```python
# menampilkan lokasi berdasarkan harga
# latitude -> Y, longitude-> X
affordable_df.plot(
    kind='scatter',
    x='longitude',
    y='latitude',
    c='price',
    cmap='inferno',
    colorbar=True,
    #alpha=0.8,
    figsize=(12,8)
)
```

```python
import urllib
i = urllib.request.urlopen('https://upload.wikimedia.org/wikipedia/commons/e/ec/Neighbourhoods_New_York_City_Map.PNG')
plt.imshow(plt.imread(i), zorder=0, extent=[-74.258, -73.7, 40.49, 40.92])
ax = plt.gca()
# menampilkan lokasi berdasarkan harga
# latitude -> Y, longitude-> X
affordable_df.plot(
    ax=ax,
    zorder=1,
    kind='scatter',
    x='longitude',
    y='latitude',
    c='price',
    cmap='inferno',
    colorbar=True,
    alpha=0.8,
    figsize=(12,8)
)
```
